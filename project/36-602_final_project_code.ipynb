```python
# data preprocessing before analysis - one hot encode columns
# need to check that classes are balanced in categories before applying to 
# only run once!
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(drop = 'first')
X_train_cat = enc.fit_transform(X_train[cat_var]).toarray()
X_test_cat = enc.transform(X_test[cat_var]).toarray()
```

```python
X_train_cat.shape
```

```python
X_train_ohe = pd.DataFrame(X_train_cat, columns = reduced_list)
```

```python
X_train_final = pd.concat([X_train.drop(columns = cat_var).reset_index(drop = True), X_train_ohe.reset_index(drop=True)], axis = 1)
```

```python
X_train_final.shape
```

```python
X_test_cat
```

```python
X_test_ohe = pd.DataFrame(X_test_cat, columns = reduced_list)
```

```python
X_test_final = pd.concat([X_test.drop(columns = cat_var).reset_index(drop = True), X_test_ohe.reset_index(drop=True)], axis = 1)
```

## model 1 = logistic regression (all variables)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, plot_confusion_matrix
logreg = LogisticRegression()
logreg.fit(X_train_final, y_train)
y_pred = logreg.predict(X_test_final)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(y_test, y_pred), display_labels = logreg.classes_).plot()
plot_confusion_matrix(logreg, X_test_final, y_test)
plt.show()
```

```python
## plotting the ROC curve
## ROC plot
# Import necessary modules
from sklearn.metrics import roc_curve
from sklearn.metrics import auc

# Compute predicted probabilities: y_pred_prob
y_pred_prob = logreg.predict_proba(X_test_final)[:,1]

# Generate ROC curve values: fpr, tpr, thresholds
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)

# Plot ROC curve
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve with AUC: ' + str(auc(fpr, tpr)))
plt.show()
```

```python
# feature selection
from sklearn.feature_selection import RFE
logreg = LogisticRegression()
selector = RFE(logreg, n_features_to_select = 10, step = 1)
selector.fit(X_train_final, y_train)
selector.support_.tolist()
```

```python
## random forest example
# Import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier

# Instantiate rf
rf = RandomForestClassifier(n_estimators=25, random_state=2)
            
# Fit rf to the training set    
rf.fit(X_train_final, y_train) 

## evaluate regressor
# Import mean_squared_error as MSE

# Predict the test set labels
y_pred = rf.predict(X_test_final)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Define the dictionary 'params_rf'
params_rf = {'n_estimators':[100, 350, 500],
             'max_features':['log2', 'auto', 'sqrt'],
             'min_samples_leaf':[2, 10, 30]}

# Import GridSearchCV
from sklearn.model_selection import GridSearchCV

rf = RandomForestClassifier(random_state=2)
# Instantiate grid_rf
grid_rf = GridSearchCV(estimator=rf,
                       param_grid=params_rf,
                       cv=3,
                       verbose=1,
                       n_jobs=-1)
best_model = grid_rf.fit(X_train_final, y_train)

# Predict test set labels
y_pred = best_model.predict(X_test_final)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

## 'analysis'

XGBoost outperforms the other models considered for the analysis when analyzing the testing data, both in terms of accuracy and f1 score (harmonic mean of precision and recall) and has better or comparable AUC to all the other models considered. Recommend using this model if none are found to be better.

In terms of feature importance (i.e. How many times was this variable split on?), of the top 5 most important features we see that 3 of the 5 (daily rate, monthly income, monthly rate) are related to compensation while the other 2 (age, total working years) are related to the age of the person/how long they have been working at the company.

Reconciliation: lower payed workers are more likely to attrite (unsurprisingly), so high performers need to be fairly compensated to encourage them to stay. Older or longer standing employees could leave for a number of reasons, such as: 1) changing industry/job to get better pay, 2) disgruntled from lack of upward mobility, 3) shifting priorities (family, illness, etc.). Employees can be very valuable as well after working for mean years at the company and are poached by other organizations to do similar work.
